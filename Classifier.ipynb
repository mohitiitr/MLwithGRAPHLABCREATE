```python
train_data , test_data = pr.random_split(0.8 , seed = 0 )
```

```python
def a_great(x): return ({'great':x['great']} if  'great' in x else {'great':0})
def a_fantastic(x): return ({'fantastic':x['fantastic']} if  'fantastic' in x else {'fantastic':0})
def a_good(x): return ({'good':x['good']} if  'good' in x else {'good':0})
def a_love(x): return ({'love':x['love']} if  'love' in x else {'love':0})
def a_amazing(x): return ({'amazing':x['amazing']} if  'amazing' in x else {'amazing':0})
def a_horrible(x): return ({'horrible':x['horrible']} if  'horrible' in x else {'horrible':0})
def a_bad(x): return ({'bad':x['bad']} if  'bad' in x else {'bad':0})
def a_terrible(x): return ({'terrible':x['terrible']} if  'terrible' in x else {'terrible':0})
def a_wow(x): return ({'wow':x['wow']} if  'wow' in x else {'wow':0})
def a_hate(x): return ({'hate':x['hate']} if  'hate' in x else {'hate':0})
```

```python
pr['great'] = pr['word_count'].apply(a_great)
pr['fantastic'] = pr['word_count'].apply(a_fantastic)
pr['good'] = pr['word_count'].apply(a_good)
pr['love'] = pr['word_count'].apply(a_love)
pr['amazing'] = pr['word_count'].apply(a_amazing)
pr['horrible'] = pr['word_count'].apply(a_horrible)
pr['bad'] = pr['word_count'].apply(a_bad)
pr['terrible'] = pr['word_count'].apply(a_terrible)
pr['wow'] = pr['word_count'].apply(a_wow)
pr['hate'] = pr['word_count'].apply(a_hate)
```
